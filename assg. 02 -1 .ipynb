```python
from scipy import stats 
from scipy.stats import norm
```

 A. More employees at the processing center are older than 44 than between 38 and 44.

here we have mean = 38 and standard deviation = 6 and n  = 400

# a

```python
# p(X>44); Employees older than 44 yrs of age
employee_older_than_44=1-stats.norm.cdf(44,loc=38,scale=6)
employee_older_than_44
```

```python
employee_older_than_44*400
```

```python
# p(38<X<44); Employees between 38 to 44 yrs of age

employee_bet_38_44 = stats.norm.cdf(44,38,6)-stats.norm.cdf(38,38,6)
employee_bet_38_44
```

```python
employee_bet_38_44 * 400
```

here employees between age 38 and 44 are 136 and older than 44 are 64 ,hence statement is false 

# b

```python
# No. of employees attending training program from 400 nos. is N*P(X<30)

400*stats.norm.cdf(30,38,6)
```

hence statement is true
